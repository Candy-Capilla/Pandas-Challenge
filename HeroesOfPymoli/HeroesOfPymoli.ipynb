```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
```

```python
#Total value
total_players = purchase_data_df["SN"].nunique()
#Create data frame
total_players_df = pd.DataFrame({"Total Players": [total_players]})
total_players_df
```

```python
#
unique_items = purchase_data_df["Item ID"].nunique()
average_price = purchase_data_df["Price"].mean()
total_purchases = purchase_data_df["Purchase ID"].count()
total_revenue = purchase_data_df["Price"].sum()
purchase_analysis_df = pd.DataFrame({"Number of Unique Items": [unique_items],
                                    "Average Price": [average_price],
                                    "Total Purchases": [total_purchases],
                                    "Total Revenue": [total_revenue]})
#
purchase_analysis_df["Average Price"] = purchase_analysis_df["Average Price"].map("${:,.2f}".format)
purchase_analysis_df["Total Revenue"] = purchase_analysis_df["Total Revenue"].map("${:,.2f}".format)
purchase_analysis_df
```

```python
#Count each gender and calculate percentage
male_players = purchase_data_df.loc[purchase_data_df["Gender"]== "Male"]
male_players_count = len(male_players["SN"].unique())
male_players_percent = (male_players_count / total_players)*100
female_players = purchase_data_df.loc[purchase_data_df["Gender"]== "Female"]
female_players_count = len(female_players["SN"].unique())
female_players_percent = (female_players_count / total_players)*100
other_players = purchase_data_df.loc[purchase_data_df["Gender"]== "Other / Non-Disclosed"]
other_players_count = len(other_players["SN"].unique())
other_players_percent = (other_players_count / total_players)*100
#Create Dataframe
gender_df = pd.DataFrame([{"Gender": "Male",
                           "Total Count": male_players_count,
                           "Percentage of Players": male_players_percent},
                          {"Gender": "Female",
                           "Total Count": female_players_count,
                           "Percentage of Players": female_players_percent},
                          {"Gender": "Other / Non-Disclosed",
                           "Total Count": other_players_count,
                           "Percentage of Players": other_players_percent}], 
#Name column headers                         
columns = (["Total Count", 
            "Percentage of Players"]))
#Rename Rows
gender_df = gender_df.rename(index={0: "Male",
                                    1: "Female",
                                    2: "Other / Non-Disclosed"})
#Fix formatting for percentage column
gender_df["Percentage of Players"] = gender_df["Percentage of Players"].map("{0:,.2f}%".format)
#Print Data
gender_df
```

```python
#Count Purchases for each Gender
male_purchases = purchase_data_df.loc[purchase_data_df["Gender"] == "Male", :]
male_purchase_count = len(male_purchases)
female_purchases = purchase_data_df.loc[purchase_data_df["Gender"] == "Female", :]
female_purchase_count = len(female_purchases)
other_purchases = purchase_data_df.loc[purchase_data_df["Gender"] == "Other / Non-Disclosed", :]
other_purchase_count = len(other_purchases)
#Calculate average price for each gender
avg_male_price = purchase_data_df.loc[purchase_data_df["Gender"] == "Male", "Price"].mean()
avg_female_price = purchase_data_df.loc[purchase_data_df["Gender"] == "Female", "Price"].mean()
avg_other_price = purchase_data_df.loc[purchase_data_df["Gender"] == "Other / Non-Disclosed", "Price"].mean()
#Calculate total purchase values for each gender
total_male_value = purchase_data_df.loc[purchase_data_df["Gender"] == "Male", "Price"].sum()
total_female_value = purchase_data_df.loc[purchase_data_df["Gender"] == "Female", "Price"].sum()
total_other_value = purchase_data_df.loc[purchase_data_df["Gender"] == "Other / Non-Disclosed", "Price"].sum()
#Calculate average purchase totals per person for each gender
avg_male_purchase = (total_male_value / male_players_count)
avg_female_purchase = (total_female_value / female_players_count)
avg_other_purchase = (total_other_value / other_players_count)
#Create Data Frame
gender_purchase_df = pd.DataFrame([{"Gender": "Male",
                                    "Purchase Count": male_purchase_count,
                                    "Average Purchase Price": avg_male_price,
                                    "Total Purchase Value": total_male_value,
                                    "Avg Total Purchase per Person": avg_male_purchase},
                                   {"Gender": "Female",
                                    "Purchase Count": female_purchase_count,
                                    "Average Purchase Price": avg_female_price,
                                    "Total Purchase Value": total_female_value,
                                    "Avg Total Purchase per Person": avg_female_purchase},
                                   {"Gender": "Other / Non-Disclosed",
                                    "Purchase Count": other_purchase_count,
                                    "Average Purchase Price": avg_other_price,
                                    "Total Purchase Value": total_other_value,
                                    "Avg Total Purchase per Person": avg_other_purchase}],
#Name column Header
columns=(["Purchase Count", 
          "Average Purchase Price", 
          "Total Purchase Value", 
          "Avg Total Purchase per Person"]))
#Rename Rows
gender_purchase_df = gender_purchase_df.rename(index={0: "Male",
                                                      1: "Female",
                                                      2: "Other / Non-Disclosed"})
#Ranem Axis
gender_purchase_df = gender_purchase_df.rename_axis(index="Gender")
#Format columns
gender_purchase_df["Average Purchase Price"] = gender_purchase_df["Average Purchase Price"].map("${:,.2f}".format)
gender_purchase_df["Total Purchase Value"] = gender_purchase_df["Total Purchase Value"].map("${:,.2f}".format)
gender_purchase_df["Avg Total Purchase per Person"] = gender_purchase_df["Avg Total Purchase per Person"].map("${:,.2f}".format)
#print data
gender_purchase_df
```
